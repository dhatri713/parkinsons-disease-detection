```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
```

```python
data = pd.read_csv("synthetic_bp_data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['SBP_RollingMean_3'] = data['SBP'].rolling(window=3, min_periods=1).mean()
data['DBP_RollingMean_3'] = data['DBP'].rolling(window=3, min_periods=1).mean()
```

```python
X_train = data.drop("Anomalous", axis=1)
y_train = data["Anomalous"]
```

```python
X_train.head()
```

```python
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(X_train, y_train)
```

```python
test_data = pd.read_csv("synthetic_bp_test_data.csv")
test_data.head()
```

```python
X_test = test_data.drop("Anomalous", axis=1)
y_test = test_data["Anomalous"]
```
